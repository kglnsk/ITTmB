```python
import torch
import timm
from torch import nn
import comet_ml
import torchvision
from torchvision import transforms,models
from torch.utils.data import Dataset
import pytorch_lightning as pl
from PIL import Image
import os
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import time 
from tqdm import tqdm


from pillow_heif import register_heif_opener

register_heif_opener()
```

```python
class LizaAlertDataset(Dataset):
    def __init__(self, root_dir, csv_path=None, transform=None):
        
        self.transform = transform
        self.files = [os.path.join(root_dir, fname) for fname in os.listdir(root_dir)]
        self.targets = None
        if csv_path:
            df = pd.read_excel(csv_path)
            self.targets = df.iloc[:,1:].values.tolist()
            self.files = [os.path.join(root_dir, fname) for fname in df["Названия"].tolist() if (".ORF" not in fname) and (".ARW" not in fname) and (".RAW" not in fname)and (".HEIC" not in fname)]

    def __len__(self):
        return len(self.files)
    
    def __getitem__(self, idx):
        target = self.targets[idx] if self.targets else -1

        image_filepath = self.files[idx]

        image = Image.open(self.files[idx])
        
        if self.transform:
            image = self.transform(image)
        return image, target

# hardcode
TRAIN_DATASET = "./Итоговый датасет (copy)/"
TRAIN_CSV = "./razm.xlsx"
```

```python
train_transform = transforms.Compose(
    [
        transforms.Resize(320),
        transforms.CenterCrop(320),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]
)
dset = LizaAlertDataset(TRAIN_DATASET, TRAIN_CSV, train_transform)
```

```python
indices = list(range(len(dset)))
ind_train, ind_test = train_test_split(indices, test_size=0.2, random_state=139)

trainset = torch.utils.data.Subset(dset, ind_train)                           
testset = torch.utils.data.Subset(dset, ind_test)

batch_size = 16
num_workers = 4
train_loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                            shuffle=True, num_workers=num_workers,prefetch_factor = 8, pin_memory=True, persistent_workers=True)

test_loader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                            shuffle=False, num_workers=num_workers,prefetch_factor = 8, pin_memory=True, persistent_workers=True)
```

```python
import torch_optimizer as optim
import torchmetrics
loss_func = nn.CrossEntropyLoss()
from torchmetrics import F1Score

class LizaNet(pl.LightningModule):
    def __init__(self):
        super(LizaNet, self).__init__()
        self.accuracy = torchmetrics.Accuracy()
        self.fscore = F1Score()
        self.model_wo_fc = timm.create_model('efficientnet_b0', pretrained=True, num_classes=0)

        self.season = nn.Sequential(
            nn.Linear(in_features=1280, out_features=5),
        )
        
        self.time_of_day = nn.Sequential(
            nn.Linear(in_features=1280, out_features=4)
        )
        self.area = nn.Sequential(
            nn.Linear(in_features=1280, out_features=3)
        )
        self.airplane = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.car = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.drone = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.scuba = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.dog = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.horse = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.hugs = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )
        self.sherp = nn.Sequential(
            nn.Linear(in_features=1280, out_features=2)
        )

    def forward(self, x):
       x = self.model_wo_fc(x)
       x = torch.flatten(x, 1)
       return {
            'time_of_day': self.time_of_day(x),
            'season': self.season(x),
            'area': self.area(x),
            'airplane': self.airplane(x),
            'car': self.car(x),
            'drone': self.drone(x),
            'scuba': self.scuba(x),
            'dog': self.dog(x),
            'horse': self.horse(x),
            'hugs': self.hugs(x),
            'sherp': self.sherp(x)
        }
    
    def training_step(self, batch, batch_idx):
        image, targets = batch
        outputs = self(image)
        losses = 0
        for i, key in enumerate(outputs):
          losses += loss_func(outputs[key], targets[i])
          self.accuracy(outputs[key], targets[i])
          self.fscore(outputs[key],targets[i])
          self.log('train_fscore'+ key, self.fscore)
          self.log('train_acc_' + key, self.accuracy)
        self.log('train_loss', losses)
        return losses

    def test_step(self, batch, batch_idx):
        image, targets = batch
        outputs = self(image)
        losses = 0
        for i, key in enumerate(outputs):
          losses += loss_func(outputs[key], targets[i])
          self.accuracy(outputs[key], targets[i])
          self.fscore(outputs[key],targets[i])
          self.log('test_acc_' + key, self.accuracy)
          self.log('test_f1_' + key, self.fscore)
        self.log('test_loss', losses)
        return losses   

    def validation_step(self, batch, batch_idx):
        image, targets = batch
        outputs = self(image)
        losses = 0
        for i, key in enumerate(outputs):
          losses += loss_func(outputs[key], targets[i])
          self.accuracy(outputs[key], targets[i])
          self.fscore(outputs[key],targets[i])
          self.log('val_acc_' + key, self.accuracy)
          self.log('val_f1_' + key, self.fscore)
        self.log('val_loss', losses)
        return losses

    def configure_optimizers(self):
      optimizer = torch.optim.Adam(self.parameters(),lr = 3e-5)
      return {'optimizer': optimizer}
```

```python
import numpy as np
datamodule = ImageClassificationData.from_files(
    predict_files=[
        "../Итоговый датасет (copy)/DSC_0152.jpg",
        "../Итоговый датасет (copy)/DSC_0152.jpg",

    ],
    batch_size=1,
)
predictions = trainer.predict(model, datamodule=datamodule, output="preds")
answers = []
for pred in predictions:
    probs = (pred[0].numpy())
    time_of_day = np.argmax(probs[:4])
    season = np.argmax(probs[4:9])
    area = np.argmax(probs[9:12])
    avia = np.argmax(probs[12:14])
    auto = np.argmax(probs[14:16])
    drone = np.argmax(probs[16:18])
    scuba = np.argmax(probs[18:20])
    dog = np.argmax(probs[20:22])
    horse = np.argmax(probs[22:24])
    hugs = np.argmax(probs[24:26])
    sherp = np.argmax(probs[26:28])
    answers.append([time_of_day,season,area,avia,auto,drone,scuba,dog,horse,hugs,sherp])
    

    
```
